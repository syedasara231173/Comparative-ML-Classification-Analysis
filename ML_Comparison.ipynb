**Data Preprocessing**

**Why we used mean imputation to handle missing values**

Mean imputation is a simple and widely-used method for handling missing values in continuous numerical data. This technique replaces missing values with the mean (average) of the available non-missing values in the same column.

**Symmetry:** If the data is approximately normally distributed and has no significant outliers, the mean represents the central tendency of the data well.

**Minimal bias**: For small datasets where missing values are random, using the mean is unlikely to introduce significant bias.

**Simplicity:** Mean imputation is computationally efficient and easy to implement.

**How Missing values affect the model**

**Identify categorical and numerical features.**

Missing values can lead to biased models or errors during training. Outliers may distort model learning, especially in algorithms sensitive to scale (like linear regression or k-NN). Both should be handled to ensure reliable predictions.
